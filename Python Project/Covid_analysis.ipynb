```python
# to answer the above question line grapgh is created.
grouped_df = df.groupby('year').agg({'total_cases': 'mean', 'total_deaths': 'mean'}).reset_index()

plt.figure(figsize=(8,4))
plt.plot(grouped_df['year'], grouped_df['total_cases'], color='skyblue', marker='o', linestyle='-',label='Total_cases')
plt.plot(grouped_df['year'], grouped_df['total_deaths'], color='orange',marker='o', linestyle='-',label='Total_deaths')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend()
plt.grid(True)
plt.title('Trendlines for total_deaths & total_cases')
plt.show()
```

6. What was the frequency of admission of patients in the hospitals?
7. How did the distribution of vaccinations increased till 2023?
8. How did COVID affect the reproduction rate?

```python
import matplotlib.pyplot as plt

df1['date'] = pd.to_datetime(df1['date'])

df1.set_index('date', inplace=True)

testing_df = df1[['hosp_patients', 'total_vaccinations', 'reproduction_rate']]
quaterly_df = df1.resample('Q').mean()

plt.figure(figsize=(9,6))

plt.subplot(2,2,1)
quaterly_df['hosp_patients'].plot(kind='line', color='crimson',marker='o')
plt.title('Hospital patients admitted Quarterly')
plt.xlabel('Period')
plt.ylabel('Hospital patients')
plt.grid(True) 

plt.subplot(2,2,2)
quaterly_df['total_vaccinations'].plot(kind='line', color='darkblue',marker='o')
plt.title('Total Vaccinations provided Quarterly')
plt.xlabel('Period')
plt.ylabel('Total Vaccinations')
plt.grid(True)   

plt.subplot(2,2,3)
quaterly_df['reproduction_rate'].plot(kind='line', color='lightgreen',marker='o')
plt.title('Reproduction rate Quarterly')
plt.xlabel('Period')
plt.ylabel('Reproduction rate')

plt.tight_layout()
plt.grid(True)           
plt.show()
           
```

9. Analyse total tests taken, positive patients rate and tests per case over this period of time.

```python
testing_df = df1[[ 'total_tests', 'positive_rate', 'tests_per_case']]
quarterly_testing = testing_df.resample('Q').mean()

plt.figure(figsize=(12, 8))


plt.subplot(2, 2, 1)
quarterly_testing['total_tests'].plot(kind='bar', color='blue')
plt.title('Total Tests Conducted Quarterly')
plt.xlabel('Quarter')
plt.ylabel('Total Tests')


plt.subplot(2, 2, 2)
quarterly_testing['positive_rate'].plot(kind='line', marker='o', color='green')
plt.title('Mean Positivity Rate Quarterly')
plt.xlabel('Quarter')
plt.ylabel('Positivity Rate')


plt.subplot(2, 2, 3)
quarterly_testing['tests_per_case'].plot(kind='area', color='orange')
plt.title('Mean Testing Capacity Quarterly')
plt.xlabel('Quarter')
plt.ylabel('Testing Capacity')

plt.tight_layout()
plt.show()
```

10. Find out vaccinations administered, people vaccinated and also the vaccination rate from 2020-2023.

```python
df1.columns
```

```python
vaccination_df = df1[['total_vaccinations','people_vaccinated','population']]
                 
vaccination_df['percentage_vaccinated']= (vaccination_df['people_vaccinated']/vaccination_df['population'])* 100
vaccination_df['vaccination_rate'] = vaccination_df['total_vaccinations'].diff().abs()

quaterly_df = vaccination_df.resample('Q').mean()

plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
quaterly_df['total_vaccinations'].plot(kind='bar', color='green')
plt.title('Total vaccinations administerd Quarterly')
plt.xlabel('Quarter')
plt.ylabel('Total vaccinations')

plt.subplot(2,2,2)
quaterly_df['percentage_vaccinated'].plot(kind='line', color='blue', marker='o')
plt.title('Polpulation vaccinated Quarterly')
plt.xlabel('Quarter')
plt.ylabel('Percentage vaccinated')
plt.grid(True)

plt.subplot(2,2,3)
quaterly_df['vaccination_rate'].plot(kind='area', color='orange')
plt.title('Mean Vaccination Rate Quarterly')
plt.xlabel('Quarter')
plt.ylabel('Vaccination rate')

plt.tight_layout()
plt.show()
```
